```python
import cv2
import os
import matplotlib.pyplot as plt

def load_images_from_folder(folder):
    images = []
    for file_name in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,file_name))
        if img is not None:
            images.append(img)
    return images
```

```python
images = load_images_from_folder("images")
```

```python
plt.imshow(images[2])
plt.show()
```
